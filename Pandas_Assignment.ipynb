# Null Values
    -The 'Customer' column has 100% null values so it will be droped

```python
#Drop the 'Customer' column
df.drop(columns=['Customer'], axis=1, inplace=True)
display(df)
```

##### The 'MonthYear' column has 8 of ' - - - - ' values that needs to be dropped
##### The 'Area (m2)' column has 850 of '#NV' values that will be droped

```python
#drop the empty values in 'MonthYear' column
df = df[df['MonthYear'] != ' - - - - ']
df
```

```python
#checking the column values after droping the empty rows
df['MonthYear'].unique()
```

```python
#splitting the 'MonthYear' column to 'Month' and 'Year' columns
df['Month'] = df['MonthYear'].apply(lambda x: int(x.split('.')[0]))
df['Year'] = df['MonthYear'].apply(lambda x: int(x.split('.')[1]))
df
```

```python
#drop the 'MonthYear' column
df.drop(columns=['MonthYear'], inplace=True)
df
```

```python
#drop the '#NV' values in 'Area (m2)' column
df = df[df['Area (m2)'] != '#NV']
df
```

```python
#there was an error while converting the data type of 'HoursOwn' column, had to drop the rows with the '?' value in it
df = df[df['HoursOwn'] != '?']
```

```python
int_cols = ['StoreID', 'Dept_ID', 'HoursLease', 'Sales units', 'Turnover']
float_cols = ['HoursOwn', 'Area (m2)']

df[int_cols] = df[int_cols].astype('int')
df[float_cols] = df[float_cols].astype('float')

df.dtypes
```

### Data Summary
    -'Time index' (numerical): the data has 9 different indices [1-9] 
    -'Country' (categorical): country where the store is located
    -'StoreID' (numerical): id for each store
    -'City' (categorical): city where the store is located
    -'Dept_ID' (numerical): id for each department in the store
    -'Dept.Name' (categorical): name for each department in the store
    - HoursOwn (numerical): total number of operating hours for the store’s owned departments
    -'HoursLease' (numerical): total number of operating hours for leased departments
    -'Sales units' (numerical): number of units sold in each department 
    -'Turnover' (numerical): total earnings in each department
    -'Area (m2)' (numerical): area in square meters
    -'Opening hours' (categorical): type of store opening ['Type A', 'Type B', 'Type C'] 
    -'Month' (numerical): month or recorded data
    -'Year' (numerical): year of recorded data
